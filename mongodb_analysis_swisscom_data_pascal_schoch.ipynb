### Comparison of Averages Datadownload, Number of SMS and calls

In the follwowing the data averages of datadownload, number of SMS and calls for the swiss cantons in july 2017 are compared. As a first step all three datasources are agregated individually and stored in the corresponding dataframe. As a last step the three dataframes are joined together and visualized.

#### Average of Datadownload

```python
agg_datendownload = db.datendownload.aggregate( [
    {"$project": {"timestamp_hour": "$fields.timestamp_hour",
                  "effective_bytes_down_per_canton_per_hour_gb": "$fields.effective_bytes_down_per_canton_per_hour_gb",
                  "effective_bytes_down_per_canton_per_hour_bytes": "$fields.effective_bytes_down_per_canton_per_hour_bytes",
                  "canton": "$fields.canton"
                  }},
    {"$match": {"effective_bytes_down_per_canton_per_hour_gb": {"$ne": 0} } },

    {"$group": { "_id": { "canton": "$canton",
                         "effective_bytes_down_per_canton_per_hour_gb": "$effective_bytes_down_per_canton_per_hour_gb"} }},
   { "$group": { "_id": "$_id.canton",
                 "avgDownload": { "$avg": "$_id.effective_bytes_down_per_canton_per_hour_gb" } } }
] )

df_agg_datendownload = pd.DataFrame(agg_datendownload)
```

#### Average of Number of SMS

```python
agg_sms = db.sms.aggregate( [
    {"$project": {"timestamp_hour": "$fields.timestamp_hour",
                  "number_sms_per_canton_per_hour": "$fields.number_sms_per_canton_per_hour",
                  "canton": "$fields.canton"
                  }},
    {"$match": {"number_sms_per_canton_per_hour": {"$ne": 0} } },

   {"$group": { "_id": { "canton": "$canton",
                         "number_sms_per_canton_per_hour": "$number_sms_per_canton_per_hour"} }},
   { "$group": { "_id": "$_id.canton",
                 "avgSMS": { "$avg": "$_id.number_sms_per_canton_per_hour" } } }
] )

df_agg_sms = pd.DataFrame(agg_sms)
```

#### Average of Number of calls

```python
agg_sprachanrufe = db.sprachanrufe.aggregate( [
    {"$project": {"timestamp_hour": "$fields.timestamp_hour",
                  "number_calls_per_canton_per_hour": "$fields.number_calls_per_canton_per_hour",
                  "canton": "$fields.canton"
                  }},
    {"$match": {"number_calls_per_canton_per_hour": {"$ne": 0} } },

   {"$group": { "_id": { "canton": "$canton",
                         "number_calls_per_canton_per_hour": "$number_calls_per_canton_per_hour"} }},
   { "$group": { "_id": "$_id.canton",
                 "avgCalls": { "$avg": "$_id.number_calls_per_canton_per_hour" } } }
] )

df_agg_sprachanrufe = pd.DataFrame(agg_sprachanrufe)
```

### Aggregation with Lookup

In the section 4.1 the three different collection were aggregated individually and joined with pandas.merge. In this chapter we want to try this again but this time with a single aggregation. For this the $lookup is needed. Lets try first a simple example where the collections datendownload and sms are joined based on timestamp_hour:

```python
for doc in db.datendownload.aggregate([
    {"$lookup": {
      "from" : "sms",
      "localField": "fields.timestamp_hour", 
      "foreignField": "fields.timestamp_hour", 
      "as": "sms_one"}},

    {"$match": {"sms_one.0" : {"$exists": True} }},
    {"$addFields": {
        "fields.number_sms_per_canton_per_hour": "$sms_one.fields.number_sms_per_canton_per_hour"}},
    {"$project": {"sms_one": 0}},
    {"$sort" : { "fields.number_sms_per_canton_per_hour" : 1}},
    {"$limit":  1} ] ):
    pprint(doc)
```

As we can see in the print out, the field number_sms_per_canton_per_hour, which was joined, contains a list of values which should not be the case. In order to only see a single value, we have to join with timestamp_hour and canton. This will be shown in the following example which is way more complex. Here the corresponding fields of sms and sprachanrufe will be joined to datendownload. As a second step the array lists with a single value will be converted to String. And last but not least the averages for each group will be created and vizualized in a Pandas Dataframe. The following link was used to join two collection based on two arguments. https://stackoverflow.com/questions/37086387/multiple-join-conditions-using-the-lookup-operator

```python
agg_lookup =  db.datendownload.aggregate([
    #Join field number_sms_per_canton_per_hour of collection sms to collection datendownload
    {"$lookup": {
           "from": "sms",
           "let": { "fields_timestamp_hour": "$fields.timestamp_hour", "fields_canton": "$fields.canton" },
           "pipeline": [
              { "$match":
                 { "$expr":
                    { "$and":
                       [
                         { "$eq": [ "$fields.timestamp_hour",  "$$fields_timestamp_hour" ] },
                         { "$eq": [ "$fields.canton", "$$fields_canton" ] }
                       ]
                    }
                 }
              }
           ],
           "as": "sms_one"}},

    {"$match": {"sms_one.0" : {"$exists": True} }},
    {"$addFields": {
        "fields.number_sms_per_canton_per_hour": "$sms_one.fields.number_sms_per_canton_per_hour"}},
    {"$project": {"sms_one": 0}},
    
    #Join field number_calls_per_canton_per_hour of collection sprachanrufe to collection datendownload
    {"$lookup": {
        "from": "sprachanrufe",
        "let": { "fields_timestamp_hour": "$fields.timestamp_hour", "fields_canton": "$fields.canton" },
        "pipeline": [
            { "$match":
                { "$expr":
                { "$and":
                    [
                        { "$eq": [ "$fields.timestamp_hour",  "$$fields_timestamp_hour" ] },
                        { "$eq": [ "$fields.canton", "$$fields_canton" ] }
                       ]
                    }
                 }
              }
           ],
           "as": "sprachanrufe_one"}},

    {"$match": {"sprachanrufe_one.0" : {"$exists": True} }},
    {"$addFields": {
        "fields.number_calls_per_canton_per_hour": "$sprachanrufe_one.fields.number_calls_per_canton_per_hour"}},
    {"$project": {"sprachanrufe_one": 0}},
    
    
    #Convert field number_calls_per_canton_per_hour from list with a single value to string
    { 
     "$addFields": { "fields.number_calls_per_canton_per_hour": 
                    { "$arrayElemAt": [ "$fields.number_calls_per_canton_per_hour", 0 ] } } },
    
    #Convert field number_sms_per_canton_per_hour from list with a single value to string
        { 
     "$addFields": { "fields.number_sms_per_canton_per_hour": 
                    { "$arrayElemAt": [ "$fields.number_sms_per_canton_per_hour", 0 ] } } },
    
    #Calculate Averages
    {"$group": {"_id": { "canton": "$fields.canton",
                         "number_calls_per_canton_per_hour": "$fields.number_calls_per_canton_per_hour",
                        "effective_bytes_down_per_canton_per_hour_gb":"$fields.effective_bytes_down_per_canton_per_hour_gb",
                        "number_sms_per_canton_per_hour": "$fields.number_sms_per_canton_per_hour"} }},
    {"$group": {"_id": "$_id.canton",
                 "avgCalls": { "$avg": "$_id.number_calls_per_canton_per_hour"}, 
                 "avgSMS":  { "$avg": "$_id.number_sms_per_canton_per_hour"},
                 "avgDatendownload":  { "$avg": "$_id.effective_bytes_down_per_canton_per_hour_gb"}
                } },
     ] )
df_agg_lookup = pd.DataFrame(agg_lookup)
df_agg_lookup
```

### Find min and max values of Datadownload using first and last

In the follwoing the min and max values of datadownload are visualized using the first and last statement.

```python
agg_datadownload_min_max = db.datendownload.aggregate( [
    {"$project": {"timestamp_hour": "$fields.timestamp_hour",
                  "effective_bytes_down_per_canton_per_hour_gb": "$fields.effective_bytes_down_per_canton_per_hour_gb",
                  "effective_bytes_down_per_canton_per_hour_bytes": "$fields.effective_bytes_down_per_canton_per_hour_bytes",
                  "canton": "$fields.canton"
                  }},

    {"$group": { "_id": { "canton": "$canton",
                         "effective_bytes_down_per_canton_per_hour_gb": "$effective_bytes_down_per_canton_per_hour_gb"} }},

    
    {"$sort": { "_id.effective_bytes_down_per_canton_per_hour_gb": 1 } },
    
    { "$group": { "_id": "$_id.canton",
               "maxDatadownload":  { "$last": "$_id.effective_bytes_down_per_canton_per_hour_gb" },
               "minDatadownload": { "$first": "$_id.effective_bytes_down_per_canton_per_hour_gb" },

             }},
    { "$project":
                { "_id": 0,
                  "canton": "$_id",
                  "maxDatadownload": 1,
                  "minDatadownload": 1,
                 
                }},
    
    {"$out": "datendownload_min_max"}
] )

df_agg_datadownload_min_max = pd.DataFrame(agg_datadownload_min_max)
df_agg_datadownload_min_max
```

A high fluctuation in Datadownload in GB can be seen for all cantons. With the $out command the min and max values are written in the DB into the collection "datendownload_min_max".

### Hourly Aggregation of averaged canton and data download

For the hourly aggregation it is needed to convert the parameter fields.timestamp_hour which is string to date:

```python
db.datendownload.update_many({ "fields.timestamp_hour" : { "$type": 2 } },
  [{"$set": {"fields.timestamp_hour":
             {"$dateFromString":
              {"dateString":
               "$fields.timestamp_hour" ,
               "format": "%Y-%m-%dT%H:%M:%S%z" } } } }])
```

As a second step lets create the aggregation:

```python
agg_datendownload_hourly = db.datendownload.aggregate( [
    {"$project": {"timestamp_hour": "$fields.timestamp_hour",
                  "effective_bytes_down_per_canton_per_hour_gb": "$fields.effective_bytes_down_per_canton_per_hour_gb",
                  "effective_bytes_down_per_canton_per_hour_bytes": "$fields.effective_bytes_down_per_canton_per_hour_bytes",
                  "canton": "$fields.canton"
                  }},
    {"$match": {"effective_bytes_down_per_canton_per_hour_gb": {"$ne": 0} } },
    
    {"$group": { "_id": {"$hour": "$timestamp_hour"},
                         "avgDownloadperHour": {"$avg":"$effective_bytes_down_per_canton_per_hour_gb"}}},
    {"$sort" : { "_id" : 1}}
] )

        
df_agg_datendownload_hourly = pd.DataFrame(agg_datendownload_hourly)
df_agg_datendownload_hourly
```

```python
plt.title("Hourly Average of Data Downlaod of an average canton in July 2017 [GB]")
plt.plot(df_agg_datendownload_hourly["avgDownloadperHour"], color = "blue", label= "avgDownloadperHour")
plt.legend(loc="upper left")
plt.xticks(rotation=45)
plt.subplots_adjust(bottom=0.15)
plt.show()
```

The chart above shows the hourly average Data Downlaod of Switzerland in July 2017. The the lowest point of Download can be seen at night arround 1 o'clock. The two peak hours for the Data download are arround 9 and 19 o'clock.
